```python
marks = [98,100,99,80.55,90]

average = sum(marks)/len(marks)
print(average)

highest = max(marks)
print(highest)

lowest = min(marks)
print(lowest)

for i in marks:
   if i >= 50:
      print(i, "Pass")
   else:
      print(i, "Fail")
```
